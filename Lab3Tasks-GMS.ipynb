# Intensity Transformations
Negative · Log · Power Law (Gamma)

```python

# ── 1. Negative ──────────────────────────────────── 
# Method 1: NumPy array manipulation
arr=np.array(img)
negative_arr = 255-arr
negative_img = Image.fromarray(negative_arr)
negative_img.save("task1_4_nrgative.jpg")

```

```python
# Method 2: PIL's ImageOps
from PIL import ImageOps
img_nrgative_ops = ImageOps.invert(img)
img_nrgative_ops.save("task1_4_nrgative_ops.jpg")
```

```python
# ── 2. Log transformation ──────────────────────────

# s = c · log(1 + r) --> We need to find c.
# We want the maximum output value (s) to be 255 when the maximum input value (r) is 255:
# 255 = c · log(1 + 255)
# c = 255 / log(1 + 255)

c = 255 / np.log(1 + 255)

# Apply the log transformation to each pixel
log_transformed_arr = c * np.log(1 + arr)
log_transformed_arr = Image.fromarray(np.uint8(log_transformed_arr))

log_transformed_arr.save("task1_5_transformed.jpg")
```

```python

# ── 3. Power-law / Gamma correction ───────────────
gamma = 0.5

arr_normalized = arr / 255.0
power_law_arr = np.power(arr_normalized, gamma)
power_law_arr = np.uint8(power_law_arr * 255)

power_law_img = Image.fromarray(power_law_arr)
power_law_img.save("task1_6_gamma.jpg")
```
